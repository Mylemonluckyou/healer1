```python
import pandas as pd
```

```python
dataset_url = 'http://mlr.cs.umass.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
df = pd.read_csv(dataset_url, sep=';')
df
#红葡萄酒数据
```

```python
from pandas import Series,DataFrame
```

```python
#查看葡萄酒质量分布情况
score = df.groupby("quality").agg({"fixed acidity":lambda x:len(x)})
score = score.reset_index()
score.columns = ['quality', 'count']
score
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
import matplotlib


%matplotlib inline
plt.rcParams['font.family'] = 'SimHei'
warnings.filterwarnings('ignore')
```

```python
sns.barplot(x = 'quality', y = 'count', data = score).set_title("葡萄酒质量分布情况")
plt.show()
```

```python
df['GoodWine'] = df.quality.apply(lambda x:1 if x >= 6 else 0)
df.tail()
```

```python
#数据集
x = np.array(df[df.columns[:11]])
x
```

```python
#标签集
y = np.array(df[df.columns[12]])
y
```

```python
# 划分数据集
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
x_train
```

```python
#数据归一化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

x_train
```

```python
#尝试建模
from sklearn.neighbors import KNeighborsClassifier

k = 3
clf = KNeighborsClassifier(k)

clf.fit(x_train, y_train)
```

```python
round(clf.score(x_train, y_train) * 100, 2)
```

```python
#交叉验证
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, x_train, y_train,
                                         cv = 5)
score = scores.mean()
round(score*100, 2)
```

```python
#择最优k值
ks = range(1, 100)
inSampleScores = []
crossValidationScores = []
d = {}

for k in ks:
    clf = KNeighborsClassifier(k).fit(x_train, y_train)
    inSampleScores.append(clf.score(x_train, y_train))
    scores = cross_val_score(clf, x_train, y_train, cv = 5)
    crossValidationScores.append(scores.mean())
    d[k] = scores.mean()
```

```python
p1 = plt.plot(ks, inSampleScores)
p2 = plt.plot(ks, crossValidationScores)
plt.legend(['train正确率', 'cv正确率'])
plt.show()
```

```python
best_k = sorted(d.items(), key = lambda x:x[1], reverse = True)[0][0]
best_k
```

```python
#使用到KNN分类器上
clf = KNeighborsClassifier(best_k).fit(x_train, y_train)
y_test_pred = clf.predict(x_test)
round(clf.score(x_test, y_test) * 100, 2)
```

```python
#混淆矩阵
from sklearn.metrics import confusion_matrix
cnf_matrix = confusion_matrix(y_test, y_test_pred)

cnf_matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_test_pred))
```
