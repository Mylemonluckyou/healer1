```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow as tf
```

```python
moon = plt.imread('./moonlanding.png')
moon.shape
```

```python
moon
```

```python
plt.figure(figsize=(12,9))
plt.imshow(moon,cmap = plt.cm.gray)
```

```python
filter_ = np.full(shape = [3,3],fill_value=1/9).astype(np.float32).reshape(3,3,1,1)


conv = tf.nn.conv2d(input = moon.reshape(1,474,630,1),filter=filter_,strides=[1,1,1,1],padding='SAME')

with tf.Session() as sess:
    ret = sess.run(conv)
    print(ret.shape)
    plt.figure(figsize=(12,9))
    plt.imshow(ret.reshape(474,630),cmap = 'gray')
```

```python
filter_ = np.array([[1/16,2/16,1/16],[2/16,4/16,2/16],[1/16,2/16,1/16]]).astype(np.float32).reshape(3,3,1,1)


conv = tf.nn.conv2d(input = moon.reshape(1,474,630,1),filter=filter_,strides=[1,1,1,1],padding='SAME')

with tf.Session() as sess:
    ret = sess.run(conv)
    print(ret.shape)
    plt.figure(figsize=(12,9))
    plt.imshow(ret.reshape(474,630),cmap = 'gray')
```
